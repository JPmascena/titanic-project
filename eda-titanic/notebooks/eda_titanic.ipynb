```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('../data/raw/train.csv')
```

```python
df.head()
df.info()
df.describe()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Sobreviventes")
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
```

```python
# Separar os dois grupos
sobreviventes = df[df['Survived'] == 1]
nao_sobreviventes = df[df['Survived'] == 0]

# Plotar os histogramas
plt.figure(figsize=(10,6))
sns.histplot(sobreviventes['Age'], color='green', label='Sobreviventes', kde=True, stat="density", alpha=0.6)
sns.histplot(nao_sobreviventes['Age'], color='red', label='No Sobreviventes', kde=True, stat="density", alpha=0.6)
plt.title('Distribuio de Idade: Sobreviventes vs No Sobreviventes')
plt.xlabel('Idade')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```
